```python

def size_convert(size):
    
    size = str(size).strip()
    
    if size == "Varies with device":
        return np.nan
    elif size.endswith("M"):
        return float(size.replace("M", ""))
    elif size.endswith("k"):
        return float(size.replace("k", "")) / 1024
    else:
        return np.nan

df["Size"] = df["Size"].apply(size_convert)
```
